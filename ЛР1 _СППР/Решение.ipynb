##  Количество "кафе" в 'район Измайлово'

```python
cafe_in_izmailovo = df[(df['TypeObject'] == 'кафе') & (df['District'] == 'район Измайлово')]
cafe_count = len(cafe_in_izmailovo)
cafe_names = sorted(cafe_in_izmailovo['Name'].drop_duplicates().tolist())

print(f"Количество кафе в районе Измайлово: {cafe_count}")
print("Названия заведений:")
for name in cafe_names:
    print(f"   - {name}")
```

## Количество ресторанов, в которых SocialPrivileges = 'да' и их названия

```python
restaurants_with_privileges = df[(df['TypeObject'] == 'ресторан') & (df['SocialPrivileges'] == 'да')]
restaurants_count = len(restaurants_with_privileges)
restaurant_names = restaurants_with_privileges['Name'].tolist()

print(f"Количество ресторанов с SocialPrivileges = 'да': {restaurants_count}")
if restaurants_count > 0:
    print("Названия заведений:")
    for name in restaurant_names:
        print(f"   - {name}")
else:
    print("Таких ресторанов нет в датасете.")
```

## Количество заведений в каждом из районов

```python
district_counts = df['District'].value_counts()

print(f"\nКоличество заведений в каждом районе:")
for district, count in district_counts.items():
    print(f"   {district}: {count}")
```
